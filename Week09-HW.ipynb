2. Create a two histograms in a subplot(vertical or horizontal is fine) that show the distribution of reviews per month and number of reviews in Benlize Airbnb data

  a, How do you choose your bin size?
  
  i chose bin size as 7 as the reviews_per_month coulmn had single digit data i through it would be clear to show it on the graph as how many reviews each month has and wheather the review count is increasing or not each month.

```python
fig = plt.figure(figsize = (8,15))

#creating vertical hist 

ax1=plt.subplot(2,1,1)

#gave x-axis data as reviews_per_month
ax1.hist(df['reviews_per_month'], histtype='step', label="reviews_per_month", bins=7)
#gave y-axis data as number_of_reviews
ax1.hist(df["number_of_reviews"], histtype='step', label="number_of_reviews", bins=7)

#set x-axis name as the column name to understand easily 
ax1.set_xlabel("reviews_per_month")
#set y-axis name as the column name to understand easily 
ax1.set_ylabel("number_of_reviews")

# created horizontal hist by giving orientation

ax2=plt.subplot(2,1,2)
#gave x-axis data as reviews_per_month 
ax2.hist(df['reviews_per_month'], histtype='step', label="reviews_per_month", bins=7, orientation="horizontal")
#gave y-axis data as number_of_reviews 
ax2.hist(df["number_of_reviews"], histtype='step', label="reviews_per_month", bins=7, orientation="horizontal")

ax2.set_xlabel("reviews_per_month")
ax2.set_ylabel("number_of_reviews")

#called legend to show the different hists 
ax1.legend()
ax2.legend()


plt.show()
```

4, What is the probability that the price for an Airbnb is under $50?

    a, What do you notice about there records?
      
      Noticed that the dataframe have different prices for each record according to the minimum_nights spent, 
      they have high price rates and low prices.
  
    b, What distribution does this data most closely fit?
    
       I used normal distribution as i found that it is easy to find statistical inference and the probability            never be 0 in this distribution so according to our dataframe prices never be 0. i also found that this            distrubtion is symmetrical and it shows mean and standard deviation at one place.

```python
#Calculated mean for price column 
P_mean = df['price'].mean()
P_mean
```

```python
#Calculated standard deviation
P_std = np.std(df['price'])
P_std
```

```python
#Calculated probability for price which is under $50
prob_under_50 = norm.cdf(50, P_mean, P_std)

print(prob_under_50)

# the probability of price under $50 is 0.29.
```

6, What percent of Airbnb listings in Belize are available 360 days a year or more?

   a, 360 days or less?
   
   b, What distribution does this data most closely fit?
   
   I think normal distribution closely fit this distribution. as normal distribution is symmetrical and it is easly to calculate mean, std. 

```python
# Counted total of Airbnb listings 
Airbnb_count = df['availability_365'].count()
Airbnb_count
```

```python
# calculated percent for Belize which are available for 360 days or more , here i have used greater than or equal sign to show the 360 days and more than 360 data
# used count of 360 days or more to see the total count and divided that with total Aitbnb count to find the %
Per_Airbnb_360_r_more= df[df['availability_365']>=360].count() / Airbnb_count * 100
print(Per_Airbnb_360_r_more)

#As per the output i see that there is 39.5% availability for 360 days a year or more.
```

# a, 

```python
# Took above data and calculated percent for 360 days or less
Per_Airbnb_360_r_less= df[df['availability_365']<=360].count() / Airbnb_count * 100
print(Per_Airbnb_360_r_less)

#As per the output i see that there is 61.5% availability for 360 days or less.
```
